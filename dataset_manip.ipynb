```python
from datasets import load_dataset

ds = load_dataset("tau/commonsense_qa")
```

```python
# from collections import Counter

# Counter(ds['validation']['answerKey'])
ds['validation']['choices']
```

```python
TEMPLATE = "Question: {question}\nOptions:\n{options}\n\nPlease answer with a single letter only. 'A', 'B', 'C', 'D' or 'E'."
def form_prompt(entry):
    options = "\n".join(f"{label}: {text}" for label, text in zip(entry['choices']['label'], entry['choices']['text']))
    return TEMPLATE.format(question=entry['question'], options=options)

print(form_prompt(ds['validation'][0]))
```
